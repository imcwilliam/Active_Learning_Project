```python
import numpy as np
import matplotlib.pyplot as plt
EasyTestPerf = np.load("Test_Performance_MNIST_Various_CompBootstrap_Final_Easy_2.npy")
HardTestPerf = np.load("Test_Performance_MNIST_Various_CompBootstrap_Final_Hard_2.npy")
```

```python
NumComplete = 10
EasyTestPerf = EasyTestPerf[0:3,:,0:NumComplete]
HardTestPerf = HardTestPerf[[0,1,2,-1],:,0:NumComplete]
```

```python
TestPerf = np.append(HardTestPerf,EasyTestPerf,axis=0)
```

```python
TestPerf.shape
```

```python
TestPerf_Mean = np.mean(TestPerf,axis=-1)
TestPerf_StE = np.std(TestPerf,axis=-1)/np.sqrt(10)
```

```python
TestPerf_Mean
```

```python
TestPerf_StE
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(13, 13))
ax.errorbar(np.array([1,2,3,4,5,6,7],dtype=int),TestPerf_Mean[[0,1,2,4,5,6,3],1],
            yerr = TestPerf_StE[[0,1,2,4,5,6,3],1],marker ='.',linestyle='none',markersize=15,capsize = 8)
my_xticks = ['DSC\nHard', 'DTC\nHard','DSTC\nHard','DSC\nEasy', 'DTC\nEasy','DSTC\nEasy','Uniform']
thisplot = plt.gca()
thisplot.set_xlim(0.5,7.5)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.title('MNIST Dynamic Active Curriculum Experiments: \nAverage Test Set Accuracy with Standard Errors',fontsize = 25)
plt.xlabel('\nTraining Method',fontsize = 25)
plt.ylabel('Test Set Accuracy (%)',fontsize = 25)
ax.axhline(y=TestPerf_Mean[3,1],linestyle='--')
ax.axvline(x=6.5,linestyle='-',c='grey',alpha=0.7)
ax.axvline(x=3.5,linestyle='-',c='grey',alpha=0.7)
plt.xticks(np.array([1,2,3,4,5,6,7],dtype=int), my_xticks,fontsize=20)
plt.show()
```

```python
fig.savefig('MNIST_Dynamic_Results.png')
```
