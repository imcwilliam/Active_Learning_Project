```python
import numpy as np
import matplotlib.pyplot as plt
EasyTestPerf = np.load("Test_Performance_Geo_Various_Easy.npy")
HardTestPerf = np.load("Test_Performance_Geo_Various_Hard.npy")
```

```python
NumComplete = 10
EasyTestPerf = EasyTestPerf[1:2,:,0:NumComplete]
HardTestPerf = HardTestPerf[:,:,0:NumComplete]
```

```python
TestPerf = np.append(HardTestPerf,EasyTestPerf,axis=0)
```

```python
TestPerf_Mean = np.mean(TestPerf,axis=-1)
TestPerf_StE = np.std(TestPerf,axis=-1)/np.sqrt(10)

BAC_Mean = np.array([0.39766429, 0.84005])
BAX_StE = np.array([0.00670788 ,0.00267679])

BAC_Mean =BAC_Mean.reshape([1,2])
BAX_StE =BAX_StE.reshape([1,2])

TestPerf_Mean = np.append(TestPerf_Mean,BAC_Mean,axis=0)
TestPerf_StE = np.append(TestPerf_StE,BAX_StE,axis=0)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(13, 13))
ax.errorbar(np.array([1,2,3,4,5,6],dtype=int),TestPerf_Mean[[0,1,2,-1-1,-1-2,-1],1],yerr = TestPerf_StE[[0,1,2,-1-1,-1-2,-1],1],marker ='.',linestyle='none',markersize=15,capsize = 8)
my_xticks = ['DSC\nHard', 'DTC\nHard','DSTC\nHard','DTC\nEasy','Uniform','BAC\nEasy']
thisplot = plt.gca()
thisplot.set_xlim(0.5,6.5)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.title('Geometric Shapes Dynamic Active Curriculum Experiments: \nAverage Test Set Accuracy with Standard Errors',fontsize = 25)
plt.xlabel('\nTraining Method',fontsize = 25)
plt.ylabel('Test Set Accuracy (%)',fontsize = 25)
ax.axhline(y=TestPerf_Mean[-1-2,1],linestyle='--')
ax.axvline(x=5.5,linestyle='-',c='grey',alpha=0.7)
ax.axvline(x=3.5,linestyle='-',c='grey',alpha=0.7)
ax.axvline(x=4.5,linestyle='-',c='grey',alpha=0.7)
plt.xticks(np.array([1,2,3,4,5,6],dtype=int), my_xticks,fontsize=20)
plt.show()
```

```python
fig.savefig('GeoShapes_Dynamic_Results_STE.png')
```
